```python
data["Hours_Studied"] = data["Hours_Studied"].fillna(data["Hours_Studied"].median())
data["Hours_Studied"].head()

```

```python
print("Teacher_Quality nulls before:")
print(data["Teacher_Quality"].isnull().sum())

data["Teacher_Quality"] = data["Teacher_Quality"].bfill().ffill()

print("Teacher_Quality nulls after:")
print(data["Teacher_Quality"].isnull().sum())

print(data["Teacher_Quality"].head())



print("\nDistance_from_Home nulls before:")
print(data["Distance_from_Home"].isnull().sum())

data["Distance_from_Home"] = data["Distance_from_Home"].bfill().ffill()

print("Distance_from_Home nulls after:")
print(data["Distance_from_Home"].isnull().sum())

print(data["Distance_from_Home"].head())

```

```python
print("Parental_Education_Level nulls before:")
print(data["Parental_Education_Level"].isnull().sum())

data["Parental_Education_Level"] = data["Parental_Education_Level"].fillna("Undefined")

print("Parental_Education_Level nulls after:")
print(data["Parental_Education_Level"].isnull().sum())

print(data["Parental_Education_Level"].head())
```

```python
xmin = data["Hours_Studied"].min()
xmax = data["Hours_Studied"].max()

data["Hours_St_Norm"] = (data["Hours_Studied"] - xmin) / (xmax - xmin)

print(data["Hours_St_Norm"].head())

plt.boxplot(data["Hours_St_Norm"])
plt.title("Box Plot of Hours_St_Norm")
plt.show()
```
